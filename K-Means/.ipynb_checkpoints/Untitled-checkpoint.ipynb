```python
import pandas as pd
import numpy as np
import random
import math

# Load data from csv file
crime_df = pd.read_csv('crime_data.csv', index_col=0)

# Get the number of clusters from user input
k = int(input("Enter the number of clusters: "))

# Define a function to calculate the Manhattan distance between two points
def manhattan_distance(x1, x2):
    return np.sum(np.abs(x1 - x2))

# Define a function to initialize centroids randomly
def initialize_centroids(data, k):
    n_features = data.shape[1]
    centroids = np.zeros((k, n_features))
    for i in range(k):
        centroid = data.sample().to_numpy()
        centroids[i] = centroid
    return centroids

# Define a function to assign data points to the nearest centroid
def assign_to_clusters(data, centroids):
    n_samples = data.shape[0]
    distances = np.zeros((n_samples, k))
    for i in range(k):
        distances[:, i] = np.apply_along_axis(manhattan_distance, 1, data, centroids[i])
    cluster_labels = np.argmin(distances, axis=1)
    return cluster_labels

# Define a function to update the centroids based on the mean of the data points in each cluster
def update_centroids(data, cluster_labels):
    centroids = np.zeros((k, data.shape[1]))
    for i in range(k):
        cluster_points = data[cluster_labels == i]
        if len(cluster_points) > 0:
            centroids[i] = np.mean(cluster_points, axis=0)
        else:
            centroids[i] = data.sample().to_numpy()
    return centroids

# Initialize centroids randomly
centroids = initialize_centroids(crime_df, k)

# Loop until convergence or maximum number of iterations is reached
max_iterations = 100
for i in range(max_iterations):
    # Assign data points to clusters
    cluster_labels = assign_to_clusters(crime_df, centroids)
    
    # Update centroids
    new_centroids = update_centroids(crime_df, cluster_labels)
    
    # Check for convergence
    if np.allclose(new_centroids, centroids):
        break
    
    # Update centroids for next iteration
    centroids = new_centroids

# Add cluster labels to original data frame
crime_df['cluster'] = cluster_labels

# Compute the Manhattan distances between each point and its centroid
distances = np.zeros((crime_df.shape[0], k))
for i in range(k):
    cluster_points = crime_df[crime_df['cluster'] == i].to_numpy()[:, :-1]
    centroid = centroids[i]
    distances[:, i] = np.apply_along_axis(manhattan_distance, 1, cluster_points, centroid)
crime_df['distance'] = distances.min(axis=1)

# Identify outliers (points with a distance greater than the mean distance plus three times the standard deviation)
outlier_threshold = crime_df['distance'].mean() + 3 * crime_df['distance'].std()
outliers = crime_df[crime_df['distance'] > outlier_threshold]

# Print the clusters and outlier records
for i in range(k):
    cluster = crime_df[crime_df['cluster'] == i]
    print(f"Cluster {i+1}:")
    print(cluster.index.tolist())
    print()
if len(outliers) > 0:
    print("Outliers:")
    print(outliers)
else:
    print("No outliers found.")
```

```python
import numpy as np
import pandas as pd


def euclidean_distance(x1, x2):
    """Compute the Euclidean distance between two points."""
    return np.sqrt(np.sum((x1 - x2) ** 2))


def k_means(X, k, max_iters=100, random_state=42):
    np.random.seed(random_state)

    # Initialize centroids randomly
    centroids = X[np.random.choice(X.shape[0], k, replace=False), :]

    # Initialize cluster assignments and distances
    clusters = np.zeros(X.shape[0])
    distances = np.zeros(k)

    for i in range(max_iters):
        # Assign each point to the nearest centroid
        for j in range(X.shape[0]):
            for c in range(k):
                distances[c] = euclidean_distance(X[j], centroids[c])
            clusters[j] = np.argmin(distances)

        # Update centroids to be the mean of the assigned points
        for c in range(k):
            assigned_points = X[clusters == c]
            if assigned_points.shape[0] > 0:
                centroids[c] = np.mean(assigned_points, axis=0)

    # Find outliers based on distance to their centroid
    outlier_indices = []
    for c in range(k):
        assigned_points = X[clusters == c]
        distances = np.zeros(assigned_points.shape[0])
        centroid = centroids[c]
        for j in range(assigned_points.shape[0]):
            distances[j] = euclidean_distance(assigned_points[j], centroid)
        median_distance = np.median(distances)
        outlier_indices.extend(np.where(distances > 3 * median_distance)[0])

    # Return cluster assignments and centroids
    return clusters, centroids, outlier_indices
# Load data from Excel file
df = pd.read_excel('Review_ratings.xlsx')

# Extract ratings columns
X = df.iloc[:, 1:].values

# Run k-means clustering
k = 4  # number of clusters
clusters, centroids, outlier_indices = k_means(X, k)

# Print cluster assignments and outlier indices
for c in range(k):
    print(f'Cluster {c}: {np.where(clusters == c)[0]}')
print(f'Outliers: {outlier_indices}')
```

```python
# UNQ_C1
# GRADED FUNCTION: find_closest_centroids

def find_closest_centroids(X, centroids):
    """
    Computes the centroid memberships for every example
    
    Args:
        X (ndarray): (m, n) Input values      
        centroids (ndarray): k centroids
    
    Returns:
        idx (array_like): (m,) closest centroids
    
    """

    # Set K
    K = centroids.shape[0]

    # You need to return the following variables correctly
    idx = np.zeros(X.shape[0], dtype=int)

    ### START CODE HERE ###
    for i in range(X.shape[0]):
          # Array to hold distance between X[i] and each centroids[j]
        distance = [] 
        for j in range(centroids.shape[0]):
            norm_ij = np.linalg.norm(X[i] - centroids[j])
            distance.append(norm_ij)

        idx[i] = np.argmin(distance)

    ### END CODE HERE ###
    
    return idx
```

```python
# UNQ_C2
# GRADED FUNCTION: compute_centpods

def compute_centroids(X, idx, K):
    """
    Returns the new centroids by computing the means of the 
    data points assigned to each centroid.
    
    Args:
        X (ndarray):   (m, n) Data points
        idx (ndarray): (m,) Array containing index of closest centroid for each 
                       example in X. Concretely, idx[i] contains the index of 
                       the centroid closest to example i
        K (int):       number of centroids
    
    Returns:
        centroids (ndarray): (K, n) New centroids computed
    """
    
    # Useful variables
    m, n = X.shape
    
    # You need to return the following variables correctly
    centroids = np.zeros((K, n))
    
    ### START CODE HERE ###
    for k in range(K):   
        points = X[idx == k]  
        centroids[k] = np.mean(points, axis = 0)

    ### END CODE HERE ## 
    
    return centroids
```

```python

def run_kMeans(X, initial_centroids, max_iters=10, plot_progress=False):
    """
    Runs the K-Means algorithm on data matrix X, where each row of X
    is a single example
    """
    
    # Initialize values
    m, n = X.shape
    K = initial_centroids.shape[0]
    centroids = initial_centroids
    previous_centroids = centroids    
    idx = np.zeros(m)
    
    # Run K-Means
    for i in range(max_iters):
        
        #Output progress
        print("K-Means iteration %d/%d" % (i, max_iters-1))
        
        # For each example in X, assign it to the closest centroid
        idx = find_closest_centroids(X, centroids)
            
        # Given the memberships, compute new centroids
        centroids = compute_centroids(X, idx, K)
    return centroids, idx
```

```python
import numpy as np
import pandas as pd
from random import randint

def manhattan_distance(x1, x2):
    return np.sum(np.abs(x1 - x2))

def initialize_centroids(X, k):
    n_samples, n_features = X.shape
    centroids = np.zeros((k, n_features))
    for i in range(k):
        centroid = X[randint(0, n_samples - 1)]
        centroids[i] = centroid
    return centroids

def kmeans(X, k, max_iters=100):
    centroids = initialize_centroids(X, k)
    for i in range(max_iters):
        clusters = [[] for _ in range(k)]
        for sample_idx, sample in enumerate(X):
            distances = [manhattan_distance(sample, centroid) for centroid in centroids]
            closest_centroid_idx = np.argmin(distances)
            clusters[closest_centroid_idx].append(sample_idx)

        prev_centroids = centroids
        centroids = np.zeros((k, X.shape[1]))
        for cluster_idx, cluster in enumerate(clusters):
            if len(cluster) > 0:
                cluster_mean = np.mean(X[cluster], axis=0)
                centroids[cluster_idx] = cluster_mean

        if np.all(prev_centroids == centroids):
            break

    return centroids, clusters

def detect_outliers(X):
    q1, q3 = np.percentile(X, [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    outliers = [i for i in range(len(X)) if X[i] < lower_bound or X[i] > upper_bound]
    return outliers

# load data from csv file
data = pd.read_csv('crime_data.csv')

# normalize the data
data = (data - data.mean()) / data.std()

# get the number of clusters from the user
k = int(input('Enter the number of clusters: '))

# perform k-means clustering
X = data.values
centroids, clusters = kmeans(X, k)

# detect outliers
outlier_indices = []
for i in range(X.shape[1]):
    outliers = detect_outliers(X[:, i])
    outlier_indices.extend(outliers)

# remove duplicates from outlier_indices
outlier_indices = list(set(outlier_indices))

# show results
for i, cluster in enumerate(clusters):
    states = data.index[cluster]
    print(f'Cluster {i + 1}: {states}')

print(f'\nOutlier indices: {outlier_indices}')
```

```python
import numpy as np
import pandas as pd

def distance(x1, x2):
    return np.sum(np.abs(x1 - x2))

def kmeans(X, k, max_iters=100):
    # randomly initialize centroids
    idx = np.random.choice(X.shape[0], k, replace=False)
    centroids = X[idx, :]
    
    for i in range(max_iters):
        # assign points to clusters
        clusters = [[] for _ in range(k)]
        for x in X:
            distances = [distance(x, c) for c in centroids]
            cluster_idx = np.argmin(distances)
            clusters[cluster_idx].append(x)
        
        # recompute centroids
        for cluster_idx, cluster in enumerate(clusters):
            if cluster:
                cluster_mean = np.mean(cluster, axis=0)
                centroids[cluster_idx] = cluster_mean
            else:
                centroids[cluster_idx] = X[np.random.choice(X.shape[0])]
    
    return centroids, clusters

data = pd.read_csv("crime_data.csv")

# get user input for k
k = int(input("Enter the number of clusters (k): "))

# perform k-means clustering
X = data.values
centroids, clusters = kmeans(X, k)

# detect outliers
outliers = [x for x in X if not any(np.array_equal(x, c) for c in centroids)]

# print clusters
for i, cluster in enumerate(clusters):
    print(f"Cluster {i+1}: {data.index[cluster]}")

# print outliers
if outliers:
    print("Outliers:")
    for outlier in outliers:
        print(data.index[np.where(X == outlier)[0][0]])
else:
    print("No outliers found.")
```

```python
import numpy as np
import pandas as pd


def initialize_centroids(data, k):
    centroids = np.zeros((k, data.shape[1]-1))
    for i in range(k):
        centroids[i] = data.sample().iloc[:, 1:5].values
    return centroids


def manhattan_distance(x, y):
    return np.sum(np.abs(x - y))


def assign_clusters(data, centroids):
    distances = np.zeros((data.shape[0], centroids.shape[0]))
    for i in range(data.shape[0]):
        for j in range(centroids.shape[0]):
            distances[i, j] = manhattan_distance(data.iloc[i, 1:5], centroids[j])
    return np.argmin(distances, axis=1)


def update_centroids(data, cluster_assignments, k):
    centroids = np.zeros((k, data.shape[1]-1))
    for j in range(k):
        centroids[j] = np.mean(data.iloc[:, 1:5][cluster_assignments == j], axis=0)
    return centroids


def detect_outliers(distances):
    q1, q3 = np.percentile(distances, [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    outlier_indices = np.where(np.logical_or(distances < lower_bound, distances > upper_bound))
    return outlier_indices[0]


def kmeans(data, k, max_iter=100):
    centroids = initialize_centroids(data, k)
    cluster_assignments = np.zeros(data.shape[0])
    prev_cluster_assignments = None
    for iter in range(max_iter):
        cluster_assignments = assign_clusters(data, centroids)
        centroids = update_centroids(data, cluster_assignments, k)
        if prev_cluster_assignments is not None and np.array_equal(prev_cluster_assignments, cluster_assignments):
            break
        prev_cluster_assignments = np.copy(cluster_assignments)
    distances = np.zeros((data.shape[0], k))
    for i in range(data.shape[0]):
        for j in range(k):
            distances[i, j] = manhattan_distance(data.iloc[i, 1:5], centroids[j])
    outlier_indices = detect_outliers(distances)
    return cluster_assignments, centroids, outlier_indices



# Load data
data = pd.read_csv('crime_data.csv')

# Define k
k = int(input("Enter the number of clusters: "))

# Perform k-means algorithm
cluster_assignments, centroids, outlier_indices = kmeans(data, k)

# Print results
for j in range(k):
    print("Cluster ", j+1, ": ", data.iloc[cluster_assignments == j,0])
outlier_states = data.iloc[outlier_indices, 0].tolist()
print("Outlier states: ", outlier_states)
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

# Load data
data = pd.read_csv('crime_data.csv')

# Define k
k = int(input("Enter the number of clusters: "))

# Initialize k-means model
model = KMeans(n_clusters=k)

# Fit model to data
model.fit(data.iloc[:, 1:5])

# Get cluster assignments and distances
cluster_assignments = model.predict(data.iloc[:, 1:5])
distances = model.transform(data.iloc[:, 1:5])

# Identify outliers
q1 = np.percentile(np.sum(distances, axis=1), 25)
q3 = np.percentile(np.sum(distances, axis=1), 75)
iqr = q3 - q1
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
outlier_indices = np.where(np.logical_or(distances < lower_bound, distances > upper_bound))
outlier_states = data.iloc[outlier_indices[0], 0].tolist()

# Print results
for j in range(k):
    print("Cluster ", j+1, ": ", data.iloc[cluster_assignments == j, 0].tolist())
print('=========================================')
print("Outlier states: ", set(outlier_states))
```
